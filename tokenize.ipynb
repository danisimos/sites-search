```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
import re

tokens = []
with open("выкачка.html", encoding='utf-8') as file:
    content = file.read()
    TAG_RE = re.compile(r'<[^>]+>')
    TAG_NUM = re.compile(r'\d+')
    TAG_UP_RU = re.compile(r'([А-Я][а-я]+[А-Я][а-я]+)')
    TAG_UP_RU2 = re.compile(r'([А-Я][а-я]+[А-Я]+)')
    TAG_UP_RU3 = re.compile(r'([а-я]+[А-Я]+[а-я]+)')
    content = TAG_RE.sub('', content)
    content = TAG_NUM.sub('', content)
    content = TAG_UP_RU.sub('', content)
    content = TAG_UP_RU2.sub('', content)
    content = TAG_UP_RU3.sub('', content)
    tokenizer = nltk.RegexpTokenizer(r"\w+")
    tokens = tokenizer.tokenize(content)
    print(tokens)
```

```python
import  pymorphy2  

def pos(word, morth=pymorphy2.MorphAnalyzer()):
    return morth.parse(word)[0].tag.POS
def pos_inflect(word, morth=pymorphy2.MorphAnalyzer()):
    return morth.parse(word)[0].lexeme

remove_pos = {'INTJ', 'PRCL', 'CONJ', 'PREP'}  # remove
tokens = [token for token in tokens if ((pos(token) not in remove_pos) and (pos(token) is not None) and (len(token) != 1))]

print(tokens)
```

```python
print(len(tokens))
tokens = list(set(tokens)) # дубликаты
print(len(tokens))
```

```python
with open("task2_result/lemmas.txt", "a", encoding="utf-8") as lemmasFile:
    for token in tokens:
        lemmasFile.write("\n" + token + ": ")
        for lemma in pos_inflect(token):
            lemmasFile.write(lemma.word + ", ")
```

```python
with open("task2_result/tokens.txt", "a", encoding="utf-8") as tokensFile:
    for token in tokens:
        tokensFile.write(token + "\n")
```
